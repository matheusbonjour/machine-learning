```python
data_path = '/kaggle/input/waverys-swsa-2018-2020/'
files = os.listdir(data_path)
files = [f for f in files if 'wav' in f]
files = [data_path + f for f in files]

waverys = xr.open_mfdataset(files, combine='by_coords')


```

**The function xr.open_mfdataset() in Xarray is used to read multiple NetCDF files into a single dataset. 
This is particularly useful when handling time-series data stored across multiple files. In the example, files matching a specific pattern are selected, and Xarray combines them 'by_coords', ensuring consistency in coordinates across files. This allows for seamless analysis of large datasets split into multiple files,effectively managing and merging them into a cohesive data structure for further operations.**
